````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20,8)
plt.style.use('ggplot')
````

````python
x = np.linspace(-np.pi, np.pi, 100)
y1 = np.cos(x)
y2 = np.tan(x)
````

````python
plt.plot(x, y1, label='sen(x)')
plt.plot(x, y2, label='cos(x)')
plt.title('Funciones de ejemplo')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
````
